```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import math
from scipy import stats
from pathlib import Path
from statistics import mean

import matplotlib
# matplotlib.rcParams['pdf.fonttype'] = 42
# matplotlib.rcParams['ps.fonttype'] = 42
import matplotlib.pyplot as plt
# plt.rc('font', family='Times New Roman')

# # plot set start
palette = sns.color_palette("tab10")
# sns.set_theme(font='Times New Roman', font_scale=1.5)
# sns.set_style("white")
```

```python
def showResults(resultFilePath):
    doplot = True

    savePath = Path(f'./img/showResults')
    savePath.mkdir(parents=True, exist_ok=True)

    resultDf = pd.read_csv(resultFilePath)

    resultDf['time'] = resultDf['time'] - resultDf['time'][0]
    resultDf['gpu'] = resultDf['gpu'] * 100
    resultDf['mem'] = resultDf['mem'] * 100
    
    # cpu
    ax = sns.lineplot(x='time', y=f'cpu', data=resultDf)
    ax.set(xlabel="Time (sec)", ylabel=f"CPU usage (percent)")

    # # plt.xticks(range(0,91,30))
    # # plt.xlim((0,90))
    plt.ylim((0,100))
    
    saveFilename = savePath/f'cpu_usage'
        
    if doplot:
        # plt.savefig(f'{saveFilename}.eps', dpi=300, pad_inches=0, bbox_inches='tight')
        plt.savefig(f'{saveFilename}.png', dpi=300, pad_inches=0, bbox_inches='tight')
        # plt.savefig(f'./img/frame_to_frame_quality/psnr_frame_pose_in_6x6_ArchVizInterior_rec20_AdditiveSynthesizer_all_pose.eps', dpi=300)
        # plt.savefig(f'./img/frame_to_frame_quality/psnr_frame_pose_in_6x6_ArchVizInterior_rec20_AdditiveSynthesizer_all_pose.png', dpi=300)
    plt.show()

    # gpu
    ax = sns.lineplot(x='time', y=f'gpu', data=resultDf)
    ax.set(xlabel="Time (sec)", ylabel=f"GPU usage (percent)")

    # # plt.xticks(range(0,91,30))
    # # plt.xlim((0,90))
    plt.ylim((0,100))
    
    saveFilename = savePath/f'gpu_usage'
        
    if doplot:
        # plt.savefig(f'{saveFilename}.eps', dpi=300, pad_inches=0, bbox_inches='tight')
        plt.savefig(f'{saveFilename}.png', dpi=300, pad_inches=0, bbox_inches='tight')
        # plt.savefig(f'./img/frame_to_frame_quality/psnr_frame_pose_in_6x6_ArchVizInterior_rec20_AdditiveSynthesizer_all_pose.eps', dpi=300)
        # plt.savefig(f'./img/frame_to_frame_quality/psnr_frame_pose_in_6x6_ArchVizInterior_rec20_AdditiveSynthesizer_all_pose.png', dpi=300)
    plt.show()

    # mem
    ax = sns.lineplot(x='time', y=f'mem', data=resultDf)
    ax.set(xlabel="Time (sec)", ylabel=f"MEM usage (percent)")

    # # plt.xticks(range(0,91,30))
    # # plt.xlim((0,90))
    plt.ylim((0,100))
    
    saveFilename = savePath/f'gpu_usage'
        
    if doplot:
        # plt.savefig(f'{saveFilename}.eps', dpi=300, pad_inches=0, bbox_inches='tight')
        plt.savefig(f'{saveFilename}.png', dpi=300, pad_inches=0, bbox_inches='tight')
        # plt.savefig(f'./img/frame_to_frame_quality/psnr_frame_pose_in_6x6_ArchVizInterior_rec20_AdditiveSynthesizer_all_pose.eps', dpi=300)
        # plt.savefig(f'./img/frame_to_frame_quality/psnr_frame_pose_in_6x6_ArchVizInterior_rec20_AdditiveSynthesizer_all_pose.png', dpi=300)
    plt.show()

    return
```

```python
if __name__ == '__main__':
    resultFilePath = Path('./Eth_mtk_demo.csv')
    showResults(resultFilePath)
    
```
